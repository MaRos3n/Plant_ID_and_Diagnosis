```python
import os
import shutil
import csv

# Paths
merged_dataset_path = '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/master_dataset'
combined_csv_path = '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/master_dataset/combined_labels.csv'

# Ensure the output directory exists
os.makedirs(merged_dataset_path, exist_ok=True)

# Function to merge datasets
def merge_datasets(dataset_name, image_dir, csv_file):
    with open(csv_file, 'r') as f:
        reader = csv.DictReader(f)
        rows = list(reader)
    
    with open(combined_csv_path, mode='a', newline='') as csv_file:
        fieldnames = ['filename', 'label']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        if os.path.getsize(combined_csv_path) == 0:  # Write header if file is empty
            writer.writeheader()

        for row in rows:
            new_filename = f"{dataset_name}_{row['filename']}"
            new_image_path = os.path.join(merged_dataset_path, new_filename)

            # Check if the file already exists
            if not os.path.exists(new_image_path):
                original_image_path = os.path.join(image_dir, row['filename'])
                shutil.copy2(original_image_path, new_image_path)
                writer.writerow({'filename': new_filename, 'label': row['label']})
                print(f"Copied {new_filename} and updated CSV.")
            else:
                print(f"Skipped {new_filename}, already exists.")

# Merge flower and leaf datasets
merge_datasets('flower', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/102_oxford_flowers', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/102_oxford_flowers/image_labels.csv')
merge_datasets('leaf', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/general_leaf_images', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/general_leaf_images/image_labels.csv')
merge_datasets('corn', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/corn_images', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/corn_images/image_labels.csv')
merge_datasets('major_crops', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/major_crop_diseases_images', '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/major_crop_diseases_images/image_labels.csv')

print("Datasets merged successfully.")
```

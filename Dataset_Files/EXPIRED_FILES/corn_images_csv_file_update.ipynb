```python
import os
import shutil
import pandas as pd

# Paths
corn_images_folder = '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/corn_images'
corn_images_csv = '/Users/maggie/Desktop/project_3/Plant_ID_and_Diagnosis/Resources/plant_dr_master_dataset/corn_images/image_labels.csv'

# Load the CSV file
df = pd.read_csv(corn_images_csv)

# Iterate over the DataFrame rows
for index, row in df.iterrows():
    old_filename = row['filename']
    new_label = row['label']

    # Construct the new filename
    new_filename = f"{new_label}_{index+1}.jpg"

    # Full paths for the old and new filenames
    old_file_path = os.path.join(corn_images_folder, old_filename)
    new_file_path = os.path.join(corn_images_folder, new_filename)

    # Rename the file
    if os.path.exists(old_file_path):
        os.rename(old_file_path, new_file_path)
        print(f"Renamed {old_filename} to {new_filename}")
    else:
        print(f"File {old_filename} not found!")

    # Update the filename in the CSV
    df.at[index, 'filename'] = new_filename

# Save the updated DataFrame back to the CSV file
df.to_csv(corn_images_csv, index=False)

print("Image renaming and CSV update completed.")
```
